# Rede Neural

```python
%matplotlib inline

import pickle as pkl
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
```

```python
import tensorflow as tf
```

```python
entrada = tf.placeholder(tf.float32, [None, 784])
```

```python
W1 = tf.Variable(tf.truncated_normal([784, 784], stddev=0.1))
b1 = tf.Variable(tf.constant(0.1, shape=[784]))

W2 = tf.Variable(tf.truncated_normal([784, 784], stddev=0.1))
b2 = tf.Variable(tf.constant(0.1, shape=[784]))

W3 = tf.Variable(tf.truncated_normal([784, 10], stddev=0.1))
b3 = tf.Variable(tf.constant(0.1,  shape=[10]))
```

```python
camada1 = tf.nn.relu(tf.matmul(entrada, W1) + b1)
camada2 = tf.nn.relu(tf.matmul(camada1, W2) + b2)
camada3 = tf.nn.relu(tf.matmul(camada2, W3) + b3)
```

```python
saida_modelo = tf.nn.softmax(camada3)
```

```python
saida_esperada = tf.placeholder(tf.float32, [None, 10])
```

```python
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=saida_esperada, logits=saida_modelo))
```

```python
train_step = tf.train.AdamOptimizer(1e-4).minimize(cross_entropy)
```

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
for iteration in range(20000):
    batch_entradas, batch_saidas = mnist.train.next_batch(50)
    sess.run(train_step, feed_dict={entrada: batch_entradas , saida_esperada: batch_saidas})
    
    if iteration % 10 == 0:
        correct_prediction = tf.equal(tf.argmax(saida_modelo, 1), tf.argmax(saida_esperada, 1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
        print("it: {}, test acc: {}".format(iteration, sess.run(accuracy, feed_dict={entrada: mnist.test.images, saida_esperada: mnist.test.labels})))
```

```python
correct_prediction = tf.equal(tf.argmax(saida_modelo, 1), tf.argmax(saida_esperada, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print(sess.run(accuracy, feed_dict={entrada: mnist.test.images, saida_esperada: mnist.test.labels}))
```
