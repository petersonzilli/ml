```python
%matplotlib inline

import pickle as pkl
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
```

```python
import tensorflow as tf
```

```python
entrada = tf.placeholder(tf.float32, [None, 784])
keep_prob = tf.placeholder(tf.float32)
```

```python
W1 = tf.Variable(tf.truncated_normal([5, 5, 1, 32], stddev=0.1))
b1 = tf.Variable(tf.constant(0.1, shape=[32]))

W2 = tf.Variable(tf.truncated_normal([5, 5, 32, 64], stddev=0.1))
b2 = tf.Variable(tf.constant(0.1, shape=[64]))

W3 = tf.Variable(tf.truncated_normal([5, 5, 64, 128], stddev=0.1))
b3 = tf.Variable(tf.constant(0.1, shape=[128]))

W4 = tf.Variable(tf.truncated_normal([4 * 4 * 128, 1024], stddev=0.1))
b4 = tf.Variable(tf.constant(0.1,  shape=[1024]))

W5 = tf.Variable(tf.truncated_normal([1024, 10], stddev=0.1))
b5 = tf.Variable(tf.constant(0.1,  shape=[10]))
```

```python
print(entrada.get_shape())
entrada_reshape = tf.reshape(entrada, [-1, 28, 28, 1])
print(entrada_reshape.get_shape())

camada_conv_1 = tf.nn.relu(tf.nn.conv2d(entrada_reshape, W1, strides=[1, 2, 2, 1], padding='SAME') + b1)
print(camada_conv_1.get_shape())

camada_conv_2 = tf.nn.relu(tf.nn.conv2d(camada_conv_1, W2, strides=[1, 2, 2, 1], padding='SAME') + b2)
print(camada_conv_2.get_shape())

camada_conv_3 = tf.nn.relu(tf.nn.conv2d(camada_conv_2, W3, strides=[1, 2, 2, 1], padding='SAME') + b3)
print(camada_conv_3.get_shape())

reshape_1 = tf.reshape(camada_conv_3, [-1, 4 * 4 * 128])
print(reshape_1.get_shape())

camada_full_1 = tf.nn.relu(tf.matmul(reshape_1, W4) + b4)
camada_full_1 = tf.nn.dropout(camada_full_1, keep_prob=keep_prob)
print(camada_full_1.get_shape())

camada_full_2 = tf.matmul(camada_full_1, W5) + b5
print(camada_full_2.get_shape())
```

```python
saida_modelo = tf.nn.softmax(camada_full_2)
```

```python
saida_esperada = tf.placeholder(tf.float32, [None, 10])
```

```python
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=saida_esperada, logits=saida_modelo))
```

```python
train_step = tf.train.AdamOptimizer(1e-4).minimize(cross_entropy)
```

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
for iteration in range(20000):
    batch_entradas, batch_saidas = mnist.train.next_batch(50)
    sess.run(train_step, feed_dict={entrada: batch_entradas , saida_esperada: batch_saidas, keep_prob:0.5})
    
    if iteration % 10 == 0:
        correct_prediction = tf.equal(tf.argmax(saida_modelo, 1), tf.argmax(saida_esperada, 1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
        print("it: {}, test acc: {}".format(iteration, sess.run(accuracy, feed_dict={entrada: mnist.test.images,
                                                                                     saida_esperada: mnist.test.labels,
                                                                                     keep_prob: 1.0})))
```

```python
correct_prediction = tf.equal(tf.argmax(saida_modelo, 1), tf.argmax(saida_esperada, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print(sess.run(accuracy, feed_dict={entrada: mnist.test.images,
                                    saida_esperada: mnist.test.labels,
                                    keep_prob: 1.0}))
```
